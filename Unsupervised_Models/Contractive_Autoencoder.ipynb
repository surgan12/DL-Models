```python
import torch.nn as nn
import torch.nn.functional as F
import os
from torch.autograd import Variable
from torchvision.utils import save_image
import torch
```

```python
class Contractive_Autoencoder(nn.Module):
    def __init__(self):
        super(Contractive_Autoencoder,self).__init__()
        self.encoder=nn.Linear(784,400)
        self.decoder=nn.Linear(400,784)
        
    def forward(self,input):
        encode=self.encoder(input)
        encode=F.sigmoid(encode)
        decode=self.decoder(encode)
        decode=F.relu(decode)
        return encode,decode
```

```python
MSE_LOSS=nn.MSELoss()

import torchvision 
import torchvision.transforms as transforms

```

```python
def contractive_loss(hidden,inputs,lam=0.01):
    #hadamard product
    hidden=hidden*(1-hidden)
    w_sum = torch.sum(Variable(W)**2, dim=1) #summing the row entries
    w_sum=w_sum.unsqueeze(1)
    contract_loss = torch.sum(torch.mm(hidden**2, w_sum), 0)
    return lam*contract_loss

assets_dir = './ass/'
if not os.path.isdir(assets_dir):
    os.mkdir(assets_dir)
transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,),(0.5,))])
realimages = torchvision.datasets.MNIST(root='./data', train=True,download=True, transform=transform)
train_loader= torch.utils.data.DataLoader(realimages, batch_size=100,shuffle=True, num_workers=2)

epochs=10

def to_img(x):
    x = 0.5 * (x + 1)
    x = x.clamp(0, 1)
    x = x.view(x.size(0), 1, 28, 28)
    return x
```

```python
import torch.optim as optim 
model=Contractive_Autoencoder()
optimizer = optim.Adam(model.parameters(), lr = 0.0001)

for epoch in  range(epochs):
    
    for i,data in enumerate(train_loader,0):
        img,_=data
        img=img.view(-1,28*28)
        hidden,output=model(img)
        W = model.state_dict()['encoder.weight']
        optimizer.zero_grad()
        loss=MSE_LOSS(output,img)+contractive_loss(hidden,W,0.01)
        loss.backward()
        print('epoch [{}/{}], loss:{:.4f}'
          .format(epoch + 1, 30, loss.data[0]))
        if epoch % 2 == 0:
            pic = to_img(output.cpu().data)
            save_image(pic, './ass/image_{}.png'.format(epoch))

    
```
